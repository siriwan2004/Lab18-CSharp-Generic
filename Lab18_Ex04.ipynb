# Lab 18 Exercise 4



## Generic method



1.สร้าง console application project

```cmd
dotnet new console --name Lab18_Ex04
```



2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
// Generic method
Generic.Print(10);
Generic.Print("Hello World");
Generic.Print(12345.678);


public static class Generic
{
    public static void Print <T> (T value)
    {
        System.Console.WriteLine(value);
    }   
}
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab18_Ex04
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง



4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3



5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab18_Ex04
```



6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง


ทุกครั้งที่เรียกใช้เมธอด Print คอมไพเลอร์จะกำหนดชนิดของ T โดยอัตโนมัติตามค่าที่ส่งเข้าไป (Type Inference) 

คำสั่ง System.Console.WriteLine(value) จะพิมพ์ค่าที่ส่งเข้ามาไปยังคอนโซล
